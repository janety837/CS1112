```python
# Calculate and draw the path of a robot given a heading.  Draw also the
# shortest paths to the nearest horizontal and vertical walls.

import random
import math
import matplotlib.pyplot as plt 
  # Use plt as an alias so we don't have to type out matplolib.pyplot every time

# Set up the window
plt.close('all')        # close all currently open figure windows
plt.figure()            # creates a new figure window
ax= plt.gca()           # gets current axes
ax.set_aspect('equal')  # unit lengths on x- and y-axis are equal (equal scaling)

# Parameter
side= 10.    # side length of the square room

# Draw walls in blue, lower-left corner at (0,0)
plt.axis((-1, side+1, -1, side+1))  # x-axis limits -1 to side+1, 
                                    # y-axis limits -1 to side+1
plt.plot([0, side],    [0, 0], 'b-', linewidth=3)  # format: Blue line of width 3
plt.plot([side, side], [0, side], 'b-', linewidth=3)
plt.plot([side, 0],    [side, side], 'b-', linewidth=3)
plt.plot([0, 0],       [side, 0], 'b-', linewidth=3)

### Do not change the code above ###
####################################
### Modify the code below ##########

# Location of Robot
xr= 8*random.random()+1
yr= 7*random.random()+1
plt.plot(xr, yr, 'ro')  # format: Red circle
plt.text(xr, yr, '  Robot')

# User-clicked point
plt.title("Click away from the robot to set the robot's heading")
plt.show()  # show the plot
point= plt.ginput(1)  # Accepts one user click; note time-out of 30 seconds
xu, yu= point[0]      # Stores x-coord of click in variable xu,
                      #        y-coord of click in variable yu
plt.plot(xu, yu, 'm+')   # format: Magenta cross
plt.title(f'You clicked at ({xu:.1f},{yu:.1f})')
#find heading
raw_inverse_tan=math.atan2(yu-yr,xu-xr)
heading=raw_inverse_tan%(2*math.pi)
#find where line connecting robot and heading intersects wall
if heading==math.pi/2:
    xs=xu #x-coord of point on side is same as xu and xr
    ys=side #y-coord of point on side is value of side
elif heading==3*math.pi/2:
    xs=xu #x-coord of point on side is same as xu and xr
    ys=0 #y-coord of point on side is 0
elif heading==0:
    xs=side #x-coord of point on side is value of side
    ys=yu #y-coord of point on side is same as yu and yr
elif heading==math.pi:
    xs=0 #x-coord of point on side is 0
    ys=yu #y-coord of point on side is same as yu and yr
else:
    m=(yu-yr)/(xu-xr)
    if heading<math.pi/2:
        #see if the line to wall intersects y=side
        ys=side
        xs=(ys-yu)/m+xu
        if xs>side:
            #line to wall must intersect x=side
            xs=side
            ys=m*(xs-xu)+yu
    elif heading<math.pi:
        #see if the line to wall intersects w/heading y=side
        ys=side
        xs=(ys-yu)/m+xu
        if xs<0:
            #line to wall must intersect x=0
            xs=0
            ys=m*(xs-xu)+yu
    elif heading<3*math.pi/2:
        #see if the line to wall intersects w/heading y=0
        ys=0
        xs=(ys-yu)/m+xu
        if xs<0:
            #line to wallmust intersect x=0
            xs=0
            ys=m*(xs-xu)+yu
    else:
        #see if the line to wall intersects w/heading y=0
        ys=0
        xs=(ys-yu)/m+xu
        if xs>side:
             #line to wall must intersect x=side
            xs=side
            ys=m*(xs-xu)+yu
# Draw line to connect the two points to a side
plt.plot([xr, xs], [yr, ys], 'r-')
#show heading
plt.title(f"Robot's heading is {heading:.1f} radians")
#draw to horizontal wall
xh=xr
if side-yr<yr:
    yh=side
else:
    yh=0
plt.plot([xr, xh], [yr, yh], 'k:')
#draw to vertical wall
yv=yr
if side-xr<xr:
    xv=side
else:
    xv=0
plt.plot([xr, xv], [yr, yv], 'm--')
####################################
### Do not change the code below ###

print('Bye')
```
